```python
import torch
import yolo
from torchvision import transforms

resize = transforms.Resize((320,320))

device = 'cuda' if torch.cuda.is_available() else 'cpu'
cuda = device == 'cuda'
dataset_test = yolo.datasets('coco', '/media/logictronix01/ML_WORKSPACE/coco2017_dataset/images/val2017', '/media/logictronix01/ML_WORKSPACE/coco2017_dataset/annotations/instances_val2017.json', train=True) # set train=True for eval
sampler_test = torch.utils.data.SequentialSampler(dataset_test)

batch_sampler_test = yolo.GroupedBatchSampler(
    sampler_test, dataset_test.aspect_ratios, 8)

data_loader_test = torch.utils.data.DataLoader(
    dataset_test, batch_sampler=batch_sampler_test, num_workers=0,  
    collate_fn=yolo.collate_wrapper, pin_memory=True)

d_test = yolo.DataPrefetcher(data_loader_test) if cuda else data_loader_test

data = next(iter(d_test))
image = data.images
targets = data.targets
image = [resize(img) for img in image]
images = torch.stack(image, dim=0)
images.shape, images.dtype, images, targets
```
