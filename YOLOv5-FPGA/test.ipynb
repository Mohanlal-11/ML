```python
import os
import glob
import re
import torch

path = 'weights/yolov5small.pth'
prefix, ext = os.path.splitext(path)
print(prefix,ext)
ckpts = glob.glob(prefix + "-*" + ext)
print(ckpts)
ckpts.sort(key=lambda x: int(re.search(r"-(\d+){}".format(ext), os.path.split(x)[1]).group(1)))
print(ckpts)
wts = torch.load(path)
print(wts.keys())
```
